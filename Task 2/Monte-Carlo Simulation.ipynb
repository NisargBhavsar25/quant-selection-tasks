# Monte-Carlo Simulation

### 1. Importing necessary Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### 2. Defining the simulation parameters

```python
timesteps = 30
starting_price = 600
volatility = 0.02
max_simulations = 5000
```

### 3. Generating the simulation output

```python
data = np.zeros((max_simulations, timesteps))
print(data)
```

```python
for i in range(max_simulations):
    price = starting_price
    for j in range(timesteps):
        data[i][j] = price
        price = np.random.normal(price, price*volatility, 1)
```

```python
print(data)
```

### 4. Calculating the Final Expected Value and it's distribution

```python
ep = data[:, -1]
print(len(ep))
print(ep)
epv = ep.mean()
```

### 5. Plotting the Scatter Plot

```python
plt.figure(figsize=(16,8))
plt.plot(data.T)
plt.xlabel('Time')
plt.ylabel('Price')
plt.text(20, 900, f"Expected Price = {epv:.2f}", fontsize="x-large")

plt.savefig(f"Scatter Plot.jpeg",bbox_inches='tight')
plt.show()
```

### 6. Plotting the Frequency Distribution Plot

```python
plt.figure(figsize=(16,8))
plt.hist(ep.T, bins=100)
plt.xlabel('Final Price')
plt.ylabel('Frequency')
plt.text(775, 150, f"Expected Price = {epv:.2f}", fontsize="x-large")

plt.savefig(f"Distribution Plot.jpeg",bbox_inches='tight')
plt.show()
```
